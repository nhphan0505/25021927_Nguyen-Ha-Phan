```python
# W4A6
a, b = map(int, input().split())
if a > b:
    a, b = b, a

def la_nt(x):
    if x < 2:
        return False
    if x == 2:
        return True
    if x % 2 == 0:
        return False
    i = 3
    while i * i <= x:
        if x % i == 0:
            return False
        i += 2
    return True

s = 0
for i in range(a, b + 1):
    if la_nt(i):
        s += i
print(s)
```

```python
# W4A7
n = int(input())
uoc = -1

while n % 2 == 0:
    uoc = 2
    n //= 2

p = 3
while p * p <= n:
    while n % p == 0:
        uoc = p
        n //= p
    p += 2

if n > 1:
    uoc = n

print(uoc)
```

```python
# W4A8
n = int(input())

def dao(x):
    r = 0
    while x > 0:
        r = r * 10 + x % 10
        x //= 10
    return r

def doi_xung(x):
    return x == dao(x)

dem = 0
while not doi_xung(n):
    n = n + dao(n)
    dem += 1

print(dem, n)
```

```python
# W4A9
n = int(input())

def khac_nhau(x):
    s = str(x)
    return len(set(s)) == len(s)

ok = False
i = 0
res = []
while i * i <= n:
    sq = i * i
    if khac_nhau(sq):
        res.append(sq)
    i += 1

if len(res) == 0:
    print("no number")
else:
    print(*res)
```

```python
# W4A10
n = int(input())

memo = {}

def dodai(x):
    if x == 1:
        return 1
    if x in memo:
        return memo[x]
    if x % 2 == 0:
        k = 1 + dodai(x // 2)
    else:
        k = 1 + dodai(3 * x + 1)
    memo[x] = k
    return k

bestx = 1
bestL = 1
for i in range(1, n + 1):
    L = dodai(i)
    if L > bestL or (L == bestL and i < bestx):
        bestL = L
        bestx = i

print(bestx, bestL)
```

```python
# W4A11
n = int(input())
dem = 0
i = 1
while i * i <= n:
    if n % i == 0:
        j = n // i
        if i % 2 == 0:
            dem += 1
        if j != i and j % 2 == 0:
            dem += 1
    i += 1
print(dem)
```

```python
# W4A13
a, b = map(int, input().split())

def tong_uoc(n):
    if n <= 1:
        return 0
    s = 1
    i = 2
    while i * i <= n:
        if n % i == 0:
            s += i
            j = n // i
            if j != i:
                s += j
        i += 1
    return s

if tong_uoc(a) == b and tong_uoc(b) == a:
    print("true")
else:
    print("false")
```

```python
# W4A15
heads, legs = map(int, input().split())
if legs % 2 != 0:
    print("invalid")
else:
    ga = (4 * heads - legs) // 2
    cho = heads - ga
    if ga < 0 or cho < 0 or 2 * ga + 4 * cho != legs:
        print("invalid")
    else:
        print(ga, cho)
```
