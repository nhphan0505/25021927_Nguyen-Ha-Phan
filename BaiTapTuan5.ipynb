```python
# W5A1
def Max_Of_Two(a, b) -> int:
    if a > b:
        return a
    return b
```

```python
# W5A2
def swap(a, b) -> tuple[int, int]:
    return b, a
```

```python
# W5A3
def isPrime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2
    return True
```

```python
# W5A4
def isPerfect(n):
    if n <= 1:
        return False
    s = 1
    i = 2
    while i * i <= n:
        if n % i == 0:
            s += i
            j = n // i
            if j != i:
                s += j
        i += 1
    return s == n
```

```python
# W5A5
def findFirst(a, k):
    for i in range(len(a)):
        if a[i] == k:
            return i
    return -1
```

```python
# W5A6
def fact(n):
    s = 1
    for i in range(1, n + 1):
        s *= i
    return s
```

```python
# W5A7
def calc(num1, num2, operat):
    if operat == '+':
        return round(num1 + num2, 2)
    if operat == '-':
        return round(num1 - num2, 2)
    if operat == '*':
        return round(num1 * num2, 2)
    if operat == '/':
        return round(num1 / num2, 2)
```

```python
# W5A8
def hamming(x, y):
    return bin(x ^ y).count("1")
```

```python
# W5A9
def sumDigits(n):
    s = 0
    while n > 0:
        s += n % 10
        n //= 10
    return s
```

```python
# W5A10
def isIsomorphic(a, b):
    if len(a) != len(b):
        return False
    m1 = {}
    m2 = {}
    for i in range(len(a)):
        c1 = a[i]
        c2 = b[i]
        if c1 in m1:
            if m1[c1] != c2:
                return False
        else:
            m1[c1] = c2
        if c2 in m2:
            if m2[c2] != c1:
                return False
        else:
            m2[c2] = c1
    return True
```
